```python
labels = ["PHY103", "CSC111", "MATH151", 
 "SALAM107", "PHY104", "MATH244", "CSC220", "CSC113", 
 "CSC304", "SALAM108", "CSC380", "CSC212", "CSC215", "CSC281"]
```

```python
 import numpy as np
lectures = np.array([8, 8, 11, 1, 4, 7, 4, 5, 2, 2, 1, 4, 2, 5])
labs = np.zeros(len(lectures), dtype="int")
labs[7] = 2; labs[-2] = 2

tutorials = np.zeros(len(lectures), dtype="int")
tutorials[5] = 4; tutorials[7] = 2; tutorials[-1]
```

```python
import matplotlib.pyplot as plt

width = 0.8
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
ax.bar(labels, lectures, width, label='Lectures')
ax.bar(labels, labs, width, bottom=lectures, label='Labs')
ax.bar(labels, tutorials, width, bottom=lectures + labs, label='Tutorials')

ax.axvspan(-0.4, 2.5, color="#DAB894", alpha=0.2, label="Lvl1") 
ax.axvspan(2.5, 7.5, color="#7E78D2", alpha=0.2, label="Lvl2") 
ax.axvspan(7.5, 13.4, color="#4cc9f0", alpha=0.2,label="Lvl3") 
ax.margins(0.0)
ax.set_ylabel('Hours')
ax.set_title('Absence')
ax.legend()

plt.show()
```
