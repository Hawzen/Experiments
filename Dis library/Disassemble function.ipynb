```python
def adder(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

def floorDivide(x, y):
    return x // y

def power(x, y):
    return x ** y

def root(x, y):
    return sqrt(x, y)
```

```python
def displayAssembly(funcs):
    for func in funcs:
        if callable(func):
            print(func)
            print(dis.dis(func), "\n\n")
```

```python
displayAssembly([adder, subtract, multiply, divide, floorDivide, power, root])
```

```python
def recursiveMultiply(x, y):
    if y == 1:
        return x
    recursiveMultiply(x + x, y - 1)
    
def recursivePower(x, y):
    if y == 0:
        return x
    recursivePower(recursiveMultiply(x, y), y-1)
```

```python
displayAssembly([recursiveMultiply, recursivePower])
```

```python
def whileloop():
    x = 0
    while (x % 100) != 0:
        print(x)
        x+=1

def forloop():
    for _ in range(10):
        pass
```

```python
displayAssembly([whileloop, forLoop])
```

```python
def bad():
    bad()

def generator():
    for i in range(10):
        yield i
```

```python
displayAssembly([bad, generator])
```
