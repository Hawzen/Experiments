```python
import os
import pandas as pd
import tabula
```

```python
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 30)
```

```python
tables = tabula.read_pdf("gradez.pdf", pages="all")
```

```python
table = pd.concat(tables)
```

```python
table = table.drop(columns=table.columns[24:])
```

```python
table[table["مسجل"] > 5].sort_values("% للراسبون", ascending=False)[:100]
```

```python
table["easy_ratio"] = table["أ+"]  / table["مسجل"]
table[table["مسجل"] > 5].sort_values("easy_ratio", ascending=False)[:100]
```

```python
grades = """
د	د+	ج	ج+	ب	ب+	أ	أ+
""".replace("\t", " ").replace("\n", "").split()

table[grades].sum().plot.bar()
```

```python
table["khalid_ratio"] = table["مسجل"] + table["الراسبون"] - table["الناجحون"]
table[table["مسجل"] > 5].sort_values("khalid_ratio", ascending=False)[:150]
```
