# Preperation

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

## Data Credits:
##       https://www.islamicfinder.org
##       https://datehijri.com/


with open("Prayer times.txt", 'r') as file:
    lines = file.read()
months = ["Jumada Al-Awwal", "Jumada Al-Akhirah", "Rajab", "Shaban", "Ramadan", "Shawwal",
          "Dhul Qadah", "Dhul Hijjah", "Muharram", "Safar", "Rabi Al Awal", "Rabi Al Akhar"]
startDayOfMonth = [(7, 36),(36, 67),(67, 96),(96, 126),(126, 155),(155, 185),(185, 214),(214, 242),
                   (242, 273),(273, 302),(302, 332),(332, 361)] # As an example, Jumada Al-Awwal starts at the seventh day
                                                                # and ends at the thirth sixith (inclusive)
    
monthDay = list(zip(months, startDayOfMonth))
```

```python
monthDay
```

```python
prayer = {}
for i, line in enumerate(lines.split("\n")):
    start = line.find(",")
    dayNum = line[0:start]
    
    prayer[dayNum] = []
    for _ in range(6):
        end = line.find(",", start+1)
        prayer[dayNum].append(line[start+1:end])
        
        start = end
```

```python
data = pd.DataFrame(prayer).T
data.columns = ["Fajr", "Sunrise", "Duhr", "Asr", "Maghrib", "Isha"] 
data
```

```python
for i in range(len(data.columns)):
    data.iloc[:,i] = pd.to_datetime(data.iloc[:,i])
pd.set_option('display.max_rows', 500)
```

# Visualization

```python
plt.style.use('seaborn-muted')
plt.subplots_adjust(hspace = 0.4)
```

```python
fig, axes = plt.subplots(6, figsize=(18, 20))
plt.setp(axes, xlabel = "Days", ylabel = "Time", xticks = [num for num in range(0, 365, 31)]) 

for i, col in enumerate(data.columns):
    axis = axes[i]
    
    axis.yaxis_date()
    axis.yaxis.set_major_formatter(mdates.DateFormatter("%I:%M %p"))
    axis.yaxis.set_major_locator(mdates.MinuteLocator(byminute = range(0, 60, 20))) # Change 2
    
    plottingData = data.loc[:, col]
    axis.plot_date(plottingData.index, plottingData, color = "#C32C6A")
    axis.set_title(col)
fig.tight_layout()
plt.show()
```

```python
fig, axis = plt.subplots(1, figsize=(15, 10))

plt.setp(axis, xlabel = "Days", ylabel = "Time", xticks = [num for num in range(-1, 365, 10)])
axis.yaxis.set_major_formatter(mdates.DateFormatter("%I:%M %p"))
axis.yaxis.set_major_locator(mdates.MinuteLocator(byminute = range(0, 60, 60))) # Change  the third parameter for better look at minutes
axis.yaxis_date()

prayerColors = ['blue', 'yellow', 'orange', 'red', 'magenta', 'purple'] # Used in coloring the plot 

for i, col in enumerate(data.columns):
    plottingData = data.loc[:, col]
    axis.plot_date(plottingData.index, plottingData, label=col, color = prayerColors[i])

colors = [[(i % 5)/5, (i % 4)/4, (i % 3)/3] for i in range(12)]

for i in range(len(monthDay)):
    interval = monthDay[i][1] # (start, end) -> (7, 36)
    month = monthDay[i][0] # String -> "Rajab"
    plt.axvspan(interval[0], interval[1], alpha=0.3, facecolor= colors[i], label=month)
    
axis.spines["top"].set_visible(False)    
axis.spines["right"].set_visible(False)    
plt.legend(loc="best")

fig.tight_layout()
plt.show()
```
